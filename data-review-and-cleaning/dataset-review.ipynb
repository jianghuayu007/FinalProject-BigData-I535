```python
import pandas as pd

def analyze_csv(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Get the total number of rows in the DataFrame
    total_rows = len(df)
    print ('There is a total of', total_rows, 'rows.')
    
    for column in df.columns:
        # Count the number of null values in the column
        null_count = df[column].isnull().sum()

        # Count the number of non-null values in the column
        non_null_count = df[column].count()

        # Calculate the percentage of null values
        null_percentage = (null_count / total_rows) * 100

        # Display the results
        print(f"\nColumn: {column}")
        print(f"Null Values: {null_count}")
        print(f"Non-Null Values: {non_null_count}")
        print(f"Null Percentage: {null_percentage:.2f}%")

input_file = 'NYPD_Shooting_Incident_Data__Historic_.csv'

analyze_csv(input_file)
```
